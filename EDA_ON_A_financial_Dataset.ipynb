# 3. Inspecting the missing values

Here, we will proceed in the following manner:
1. Check the missing values in all columns of df.
2. Checking the columns with missing values equal to or more than 50%.
3. Drop the columns having missing values equal to or more than 50%.

```python
# checking the null values in dataframe df.

df.isnull().sum()
```

```python
# checking the columns having null values more than 50%

null_values=round((df.isnull().sum()/len(df.index))*100,4)
null_values[null_values>=50]
```

Since the missing values of the columns are more than 50%, there is no resonable imputation technique to impute the missing values. Also if missing values are imputed with mean/median/mode/0, it would harm the analysis and would generate misleading results.
So, Its better to drop the columns with equal to or more than 50% missing values.

```python
# checking the number of columns having missing values equal to or more than 50%.

len(null_values[null_values>=50])
```

So there are total 41 columns having missing values >=50%

```python
# drop the columns having missing values equal to or more than 50%.

high_missing_values_columns = (null_values[null_values>=50].index)
df.drop(labels =high_missing_values_columns, axis=1, inplace=True)
```

```python
# Checking the shape of df after dropping the columns
df.shape
```

The shape of df, after dropping the columns, indicates that there are 81 columns left. This could also be verified as 81(columns left after dropping) + 41(columns having missing values >= 50%) = 122 (original number of columns in df)

### 11. Correlation Matrix

#### Here we are plotting the correlation matrix for all the numerical columns of TARGET_0 Dataframe

```python
correlation=TARGET_0.corr()
correlation
```

#### Here for many of the columns we see that the values are repeated. Hence to get rid of the repeated correlation values we follow the below steps:
* Atfirst, we convert the diagonal and the value below the diagonal of the matrix to False, False is replaced with NaN.
* Next we convert this table into a dataframe
* Now, we need to change the names of the columns that we've got in the dataframe
* Next, drop the columns having missing values as they are not required
* For better readability, we'll round off the values
* Since we see correlation as an absolute value, we are converting it into absolute value
* Next, we'll sort the correlation values

```python
# Atfirst, we convert the diagonal and the value below the diagonal of the matrix to False, False is replaced with NaN.
correlation=correlation.where(np.triu(np.ones(correlation.shape), k=1).astype(np.bool))
correlation
```

```python
#Next we convert this table into a dataframe
corrdf = correlation.unstack().reset_index()
corrdf.head()
```

```python
#Now, we need to change the names of the columns that we've got in the dataframe
corrdf.columns=['COL1','COL2','Correlation']
corrdf.head()
```

```python
#Next, drop the columns having missing values as they are not required
corrdf.dropna(subset = ['Correlation'], inplace = True)
corrdf.head()
```

```python
#For better readability, we'll round off the values
corrdf['Correlation'] = round(corrdf['Correlation'], 2)
corrdf.head()
```

```python
#Since we see correlation as an absolute value,hence we are converting it into absolute value
corrdf['Correlation'] = corrdf['Correlation'].abs()
corrdf.head()
```

```python
#Next, we'll sort the correlation values
sorted_corrdf = corrdf.sort_values(by = 'Correlation', ascending = False).head(10)
sorted_corrdf
```

### INFERENCE:
From the matrix(for TARGET_0), it is pretty clear that the columns, OBS_60_CNT_SOCIAL_CIRCLE and OBS_30_CNT_SOCIAL_CIRCLE are most correlated as they hold a correlation value=1. Also, AMT_GOODS_PRICE and AMT_CREDIT hold a correlation value of 0.99 , which is also pretty high and it can be stated that these columns are highly correlated.
AMT_GOODS_PRICE and AMT_CREDIT hold a correlation value of 0.78
